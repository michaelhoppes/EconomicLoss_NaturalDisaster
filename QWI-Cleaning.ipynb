```python
import pandas as pd
import cenpy as cenpy
```

```python
# Quaterly Work Indicators (through Q1 2018)
qwi = pd.read_csv('../project-4/data/master_qwi_2003_2018.csv')

qwi.drop(columns = ['Unnamed: 0', 'industry'], inplace= True)

qwi.drop_duplicates(inplace= True)
```

```python
# Set index to quarterly date time
qwi.index = pd.DatetimeIndex(qwi['time'].values)
g_unit = 'county:*'
g_filter = {'state': '02'}
time = 'from 2003-Q1 to 2018-Q1'
```

```python
state_codes = {
    'WA': '53', 'DE': '10', 'DC': '11', 'WI': '55', 'WV': '54', 'HI': '15',
    'FL': '12', 'WY': '56', 'NJ': '34', 'NM': '35', 'TX': '48',
    'LA': '22', 'NC': '37', 'ND': '38', 'NE': '31', 'TN': '47', 'NY': '36',
    'PA': '42', 'AK': '02', 'NV': '32', 'NH': '33', 'VA': '51', 'CO': '08',
    'CA': '06', 'AL': '01', 'AR': '05', 'VT': '50', 'IL': '17', 'GA': '13',
    'IN': '18', 'IA': '19', 'MA': '25', 'AZ': '04', 'ID': '16', 'CT': '09',
    'ME': '23', 'MD': '24', 'OK': '40', 'OH': '39', 'UT': '49', 'MO': '29',
    'MN': '27', 'MI': '26', 'RI': '44', 'KS': '20', 'MT': '30', 'MS': '28',
    'SC': '45', 'KY': '21', 'OR': '41', 'SD': '46'
}

# Reverse dictionary to map codes to states
state_codes_rev = dict((int(v),k) for k,v in state_codes.items())

qwi['state_abv'] = qwi['state'].map(state_codes_rev)

# Compile county / state codes 

# Create placeholder dataframe
county_codes = cenpy.explorer.fips_table('county', in_state = 'WA')

# Iterate through all states
for k, v in state_codes.items():
    df_1 = cenpy.explorer.fips_table('county', in_state = k)
    
    county_codes = pd.concat([county_codes, df_1])

# Combine county and state codes
county_codes[0] = county_codes[1].astype(str) + ':' + county_codes[2].astype(str)

# Drop and rename columns
county_codes.drop(labels=[1, 2, 4], axis= 1, inplace = True)

county_codes.rename(columns={0: 'code', 3: 'county'}, inplace = True)

# Set index for dictionary
county_codes.set_index(county_codes['code'], inplace = True)

county_codes.drop(labels=['code'], axis= 1, inplace = True)

# Convert dataframe to dict
county_codes_dict = county_codes.to_dict('dict')

county_codes_dict = dict((k,v) for k,v in county_codes_dict['county'].items())

# Combine county and state codes
qwi['county_codes'] = qwi['state'].astype(str) + ':' + qwi['county'].astype(str)

qwi['county_name'] = qwi['county_codes'].astype(str).map(county_codes_dict)
```
